```python
Year=[2012,2014,2017,2019,2022]
India=[58.5, 70, 89.12, 102.13, 200]
Bangladesh=[49.3,65.5, 79.45, 102,167.70]
plt.plot(Year, India, color='red')
plt.plot(Year, Bangladesh, color='green')
plt.plot(Year, India, color='green')
plt.plot(Year, Bangladesh, color='red')
plt.show()
```

```python
plt.plot(Year, India, color='green', label='India')
plt.plot(Year, Bangladesh, color='red', label='Bangladesh')
plt.legend()
plt.show()
```

```python
plt.plot(Year, India, color='green', label='India', marker='o')
plt.plot(Year, Bangladesh, color='red', label='Bangladesh', marker='o')
plt.legend()
plt.show()
```

```python
plt.plot(Year, India, color='green', label='India', marker='d')
plt.plot(Year, Bangladesh, color='red', label='Bangladesh', marker='x')
plt.legend()
plt.show()
```

```python
from matplotlib.markers import MarkerStyle
```

```python
plt.plot(Year, India, color='green', label='India', marker='d', markersize=12)
plt.plot(Year, Bangladesh, color='red', label='Bangladesh', marker='x', markersize=10)
plt.legend()
plt.show()
```

```python
plt.plot(Year, India, color='green', label='India', marker=d', markersize=12, mec='r')
plt.plot(Year, Bangladesh, color='red', label='Bangladesh', marker='x', markersize=10)
plt.legend()
plt.show()
```

```python
plt.plot(Year, India, color='green', label='India', marker='d', markersize=12, mec='r')
plt.plot(Year, Bangladesh, color='red', label='Bangladesh', marker='s', markersize=10, mec='black', mfc='blue')
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
print(x)
print(y)
```

```python
plt.scatter(x, y, c='g', label='Marks')
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.savefig('Marks.png')
plt.show()
```

```python
y = x * x
plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=10)
plt.title("2D Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 2)
plt.plot(y, x, 'g--')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'b--')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'b*')
```

```python
import pandas as pd
import seaborn as sns
```

```python
data = [1, 2, 3, 4, 5, 6, 7, 8]
df = pd.DataFrame({'Value': data})
# Create violin plot
sns.violinplot(data=df)
# Adding title and labels
plt.title("Violin Plot ")
plt.xlabel("Day")
plt.ylabel("Total")
# Displaying the plot
plt.show()
```
